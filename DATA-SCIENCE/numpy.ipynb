# Why use NumPy arrays instead of Python lists?
# - NumPy arrays are faster for numeric computing because values are stored in contiguous memory.
# - They allow vectorized operations, so arithmetic works elementwise without explicit Python loops.
# - All elements share the same data type, which makes math operations more efficient.
# - NumPy includes many built-in functions for reshaping, slicing, and linear algebra.

```python
# this is specifying the data type of the array elements.
#  By default, numpy will try to infer the data type based on the input data, 
# but you can explicitly set it using the dtype parameter when creating the array.

# create an array that will be kept in float64 bits
array_float = np.array([1.0, 2.0, 3.0], dtype=np.float64)
print(f'array with float64 data type: {array_float}')

# float64 is a data type that represents floating-point numbers using 64 bits.
# the 64 bits represent storage for the number, allowing for a wide range of values and high precision.

# create an array that will be kept in int64 bits
array_int = np.array([1, 2, 3], dtype=np.int64)
print(f'array with int64 data type: {array_int}')
```

```python
# full returns a new array of given shape and type, filled with fill_value.
f = np.full((7,7), 8, dtype=np.float64) # this will create an array of 7x7 elements, all initialized to 8.
print(f'array filled with 8: {f}')
```

```python
nums = np.arange(10) # this will create an array of 10 elements, from 0 to 9.
# print(f'arange array: {nums}')

# we could specify the start and end of the range as well
nums_range = np.arange(5, stop=15) # this will create an array of 10 elements, from 5 to 14.
# print(f'arange array with range: {nums_range}')


# we specify the step as well
nums_step = np.arange(0, 20, 2) # this will create an array of even numbers from 0 to 18.
print(f'arange array with step: {nums_step}')


# linspace - it is used to create an array of evenly spaced numbers over a specified interval.
linspace_array = np.linspace(5, 10, num = 5, dtype =np.int64) # this will create an array of 5 evenly spaced numbers between 5 and 10.
print(f'linspace array: {linspace_array}')
# the mathematical formula for linspace is:
# start + (stop - start) * (num - 1) / (num - 1)
```

```python
# changing the shape of an array

# create a 16 element array
array  = np.full((16,),2) # this will create an array of 16 elements, all initialized to the value of the original array.
print (f'array: {array}')
```

```python
# to chnage the shape
array = array.reshape(4,4) # this will change the shape of the array to 4x4.
print(f'array with new shape: {array}')
```

```python
nums  = np.arange(20) # this will create an array of 20 elements, from 0 to 19.
print(f'arange array: {nums}')

# to change the shape
nums = nums.reshape(4,5) # this will change the shape of the array to 4x5.
print(f'array with new shape: {nums}')

# why would you want to change the shape of an array - it is because sometimes we need to 
# perform operations on arrays that require them to have a specific shape, or we may want to 
# reshape an array for better visualization or to fit the requirements of a particular algorithm.
```

```python
# reshaping the reshaped array back to its original shape
nums = nums.reshape(20) # this will change the shape of the array back to its original shape.
print(f'array reshaped back to original shape: {nums}')
```

```python
np.vstack((a,b))
# vstack is used to stack arrays in sequence vertically (row wise). 
# It takes a tuple of arrays as input and returns a single array that is 
# the result of stacking the input arrays vertically. In this case,
#  it will stack array 'a' on top of array 'b', resulting in an 8x4 array.
```

```python
# Horizontal stacking
np.hstack((a,b))
# hstack is used to stack arrays in sequence horizontally (column wise). 
# It takes a tuple of arrays as input and returns a single array that is 
# the result of stacking the input arrays horizontally. In this case,
#  it will stack array 'a' on the left of array 'b', resulting in an 4x8 array.
```
